```python
import tensorflow as tf
import numpy as np
import os

model_path = "/Users/mohammedalqadda/Autonomous_Gardening_Robot_for_Home_Care/Machine_Learning/trained_model.keras"
model = tf.keras.models.load_model(model_path)

# Define the prediction function
def model_prediction(test_image):
    image = tf.keras.preprocessing.image.load_img(test_image,target_size=(256,256))
    input_arr = tf.keras.preprocessing.image.img_to_array(image)
    input_arr = np.array([input_arr]) 
    predictions = model.predict(input_arr)
    return predictions

class_names = ['Coleus_Blumei', 'Green_Pepper', 'Guldawari']
class_name_pred = ['Coleus_Blumei', 'Green_Pepper', 'Guldawari']

image_dir = "/Users/mohammedalqadda/Autonomous_Gardening_Robot_for_Home_Care/Machine_Learning/test"

correct_predictions = 0
total_images = 0

for root, dirs, files in os.walk(image_dir):
    for file_name in files:
        if file_name.endswith(".jpeg"):
            file_path = os.path.join(root, file_name)

            pred = model_prediction(file_path)
            result_index = np.argmax(pred) 
            pred_percentage = np.round(pred[0][result_index] * 100, 2)
            predicted_class = class_name_pred[result_index]

            true_class = None
            for class_name in class_names:
                if class_name in file_name:
                    true_class = class_name
                    break
            if true_class is None:
                print(f"Warning: Could not determine the true class for file {file_name}")
                continue

            if true_class == predicted_class:
                correct_predictions += 1

            total_images += 1

            print(f"Prediction for {file_name}: {predicted_class} ({pred_percentage}%) - True Class: {true_class}")

if total_images > 0:
    accuracy = (correct_predictions / total_images) * 100
    print(f"\nOverall Accuracy: {accuracy:.2f}% ({correct_predictions}/{total_images})")
else:
    print("No images found for testing.")
```
